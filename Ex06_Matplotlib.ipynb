## 데이터 시각화
- 범주형 범수 : 정해져 있는 것 ( 혈액형 )
- 범위형 범수 : 수치 데이터

#### Matplotlib
1) pyplot : 시각화, 비대화형
2) pylab : 시각화 + numpy
- 수업에선 pyplot을 사용함

```python
import matplotlib.pyplot as plt
```

### 선그래프

```python
# plot함수 - 일직선으로 되어있는 Lineplot 그리기
y = [2,4,6]
plt.plot(y) # 그래프 그리기
plt.show()  # 그래프 출력
```

```python
# x축 설정하여 그래프 그리기
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = []
for i in range(1,10):
    i = 3*i
    y.append(i)

plt.plot(x,y)
plt.show()
```

```python
# line style
# 그려지는 라인의 스타일 지정
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y,linestyle='--')
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [10, 13, 16, 20]

y2 = [10, 20, 16, 13]
plt.plot(x,y)
plt.plot(x,y2, ls = ':')  # 점선 표시
plt.show()
```

```python
# marker
plt.plot(x,y,marker='o')
plt.show()
```

```python
# linewidth , lw로 줄여씀
# 라인 두께 지정
plt.plot(x, y, linewidth = '5')
plt.show()
```

```python
plt.plot(x, y, color = 'r')
plt.plot(x, y, color = '#0000ff')
plt.show()
```

```python
# Marker color
# markerfacecolor 로 마커 색 변경 mfc 로 줄여씀
plt.plot(x, y, color = '#039BE5', marker = 'o' , markerfacecolor = 'r')
plt.show()
```

```python
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls = '--',color = 'g' ,marker ='o', markerfacecolor='g')
plt.show()
```

```python
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

plt.plot(x,y, ls = ':',color = 'r' ,marker ='s', markerfacecolor='b')
plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, linestyle = '--', color = 'b',linewidth = 5, marker='o',markersize = 10,
         markerfacecolor = 'r', markeredgecolor = 'g', markeredgewidth = 3)
plt.show()
```
